```python
import yt_dlp
import os

def download_video(url, output_path, custom_name="hello.mp4"):
    # Use video title if no custom name is provided
    out_name = f"{custom_name or '%(title)s'}.%(ext)s"

    ydl_opts = {
        'outtmpl': os.path.join(output_path, out_name)  # Specify output path and filename
    }

    with yt_dlp.YoutubeDL(ydl_opts) as ydl:
        info = ydl.extract_info(url, download=True)  # Download video
        metadata = {
            "Author": info.get("uploader"),
            "Title": info.get("title"),
            "Views": info.get("view_count")
        }

    print(f"Downloaded video: {metadata['Title']}")
    return metadata
```

```python
import os
from moviepy.editor import VideoFileClip

def video_to_images(output_video_path, output_folder_path):

    video_file_path = os.path.join(output_video_path, 'hello.mp4.webm')
    if not os.path.exists(video_file_path):
        raise FileNotFoundError(f"Video file not found at: {video_file_path}")

    clip = VideoFileClip(video_file_path)
    clip.write_images_sequence(
        os.path.join(output_folder, "frame%04d.png"), fps=0.2
    )
```

```python
def video_to_audio(newfilePath, output_audio_path):
  clip = VideoFileClip(newfilePath)
  audio = clip.audio
  audio.write_audiofile(output_audio_path)
```

```python
def audio_to_text(audio_path):
  recognizer=sr.Recognizer()
  audio=sr.AudioFile(audio_path)

  with audio as source:
    audio_data=recognizer.record(source)

    try:
      text=recognizer.recognize_google(audio_data)

    except  sr.UnknownValueError:
      print("Speech recongnition could understand the  audion")

  return text
```

```python
from llama_index.core.indices import MultiModalVectorStoreIndex
from llama_index.core import SimpleDirectoryReader
from llama_index.core import StorageContext
from llama_index.vector_stores.lancedb import LanceDBVectorStore
```

```python
text_store = LanceDBVectorStore(uri="lancedb", table_name="text_collection")
image_store = LanceDBVectorStore(uri="lancedb", table_name="image_collection")
```

```python
storage_context = StorageContext.from_defaults(
    vector_store=text_store,
    image_store=image_store
)
```

```python
documents= SimpleDirectoryReader(output_folder).load_data()
```

```python
index = MultiModalVectorStoreIndex.from_documents(
    documents,
    storage_context=storage_context
)
```

```python
retriever = index.as_retriever(similarity_top_k=1, image_similarity_top_k=5)
```

```python
qa_tmpl_str = (
    "Based on the provided information, including relevant images and retrieved from the video, "
    "accurately and precisely answer the query without any additional prior knowledge.\n"

    "--------------------\n"
    "Context: {context_str}\n"
    "Metadata for video: {metadata_str}\n"

    "---------------------------\n"
    "Query: {query_str}\n"
    "Answer: "
)
```

```python
from  llama_index.core.response.notebook_utils import display_source_node
from llama_index.core.schema import ImageNode
```

```python
def retrieve(retriever, query_str):
    retrieval_results = retriever.retrieve(query_str)

    retrieved_image = []
    retrieved_text = []

    for res_node in retrieval_results:
        if isinstance(res_node.node, ImageNode):  # Check if node is an image
            retrieved_image.append(res_node.node.metadata["file_path"])  # Collect image path
        else:
            display_source_node(res_node, source_length=200)  # Display source node for text
            retrieved_text.append(res_node.text)  # Collect text from result node

    return retrieved_image, retrieved_text
```

```python
query="Tell me something about the alphago what it was"
```

```python
img, text = retrieve(retriever, query)
```

```python
import os
from PIL import Image
import matplotlib.pyplot as plt

def plot_images(image_paths):
    images_shown = 0
    plt.figure(figsize=(16, 9))

    for img_path in image_paths:
        if os.path.isfile(img_path):
            image = Image.open(img_path)

            plt.subplot(2, 3, images_shown + 1)  # 2 rows, 3 columns, position images_shown + 1
            plt.imshow(image)
            plt.xticks([])  # Hide x-axis ticks
            plt.yticks([])  # Hide y-axis ticks

            images_shown += 1
            if images_shown >= 5:  # Limit the number of images to show
                break

    plt.show()  # Display the plot

```
